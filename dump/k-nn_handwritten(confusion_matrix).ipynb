```python
import os
import numpy as np
from imageio import imread, imsave
from skimage.feature import peak_local_max, hog
from sklearn.model_selection import  train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from skimage.transform import resize
```

```python
features_list = []
features_label = []

label_set = "abcdefghijklmnopqrstuvwxyz"

for character in label_set:
    label = character
    training_directory = './Datasets/Handwritten/' + str(label) + '/'
    for filename in os.listdir(training_directory):
        if (filename.endswith('.png')):
            image = imread(training_directory + filename)
            image = resize(image, (200, 200))
            hog_features = hog(image, orientations=12, pixels_per_cell=(16,16), cells_per_block=(1, 1))
            features_list.append(hog_features)
            features_label.append(label_set.find(label))
features = np.array(features_list)
```

```python
features = np.array(features_list)
X_train, X_test, y_train, y_test = train_test_split(features, features_label, test_size=0.3, random_state=0)
```

```python
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)
```

```python
X_predict = knn.predict(X_test)
accuracy = accuracy_score(y_test, X_predict)
print(accuracy)
```
